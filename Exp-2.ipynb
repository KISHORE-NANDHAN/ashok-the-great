# **Importing** **libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y,c="red")
plt.show()
```

```python
iris = pd.read_csv('/iris.csv')
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("sepal lenght vs sepal width")

setosa = iris[iris["Species"] == "Iris-setosa"]
versicolor = iris[iris["Species"] == "Iris-versicolor"]
virginica = iris[iris["Species"] == "Iris-virginica"]

plt.scatter(setosa["SepalLengthCm"], setosa["SepalWidthCm"], c="red", label="setosa")
plt.scatter(versicolor["SepalLengthCm"], versicolor["SepalWidthCm"], c="blue", label="versicolor")
plt.scatter(virginica["SepalLengthCm"], virginica["SepalWidthCm"], c="yellow", label="virginica")
plt.legend()
plt.show()
```

```python
plt.scatter(setosa['PetalLengthCm'],setosa['PetalWidthCm'],c="black",label="setosa")
plt.scatter(versicolor['PetalLengthCm'],versicolor['PetalWidthCm'],c="brown",label="versicolor")
plt.scatter(virginica['PetalLengthCm'],virginica['PetalWidthCm'],c="navy",label="virginica")
plt.legend()
plt.show()
```
