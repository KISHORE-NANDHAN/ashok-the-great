```python
#principal component analysis program
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

#define the 2d dataset
data = np.array([[2, 3], [3, 3], [4, 4], [5, 5]])

#create a PCA object to reduce the data to 1D
pca = PCA(n_components=1)

#fit the PCA model to the data and transform the data
data_1d = pca.fit_transform(data)

#print the transformed 1d data
print("Original 2d data: ")
print(data)

print("Transformed 1d data: ")
print(data_1d)

#optional : Visualize the original data and the pca direction
plt.scatter(data[:, 0], data[:, 1], label="Original data")
plt.quiver(pca.mean_[0], pca.mean_[1], pca.components_[0,0], pca.components_[0,1] ,angles='xy',scale_units='xy',scale=1, color='r',label='PCA direction')
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("PCA on 2d data")
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns

#load the iris dataset
iris = datasets.load_iris()

#create a dataframe with the data and correct column names
df = pd.DataFrame(iris['data'],columns=iris['feature_names'])

#display the first few rows of data frame
print(df.head())

scalar = StandardScaler()
scaled_data = pd.DataFrame(scalar.fit_transform(df))
scaled_data.head()
sns.heatmap(scaled_data.corr())
pca = PCA(n_components = 3)
pca.fit(scaled_data)
data_pca = pca.transform(scaled_data)
data_pca = pd.DataFrame(data_pca, columns = ['PC1', 'PC2', 'PC3'])
data_pca.head()

sns.heatmap(data_pca.corr())

```
