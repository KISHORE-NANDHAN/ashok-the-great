```python
data = pd.Series(data = [1000,2000,3000,4000,5000],index=['swift','jazz','civic','altis','gallardo'])
print(data)
```

```python
car_dict = {
    'swift':1000,
    'jazz':2000,
    'civic':3000,
    'altis':4000,
    'gallardo':5000
    }
car_price = pd.Series(car_dict)
car_price
```

```python
car_man_dict = {
    'swift':'Maruti',
    'jazz':'Maruti',
    'civic':'Honda',
    'altis':'Maruti',
    'gallardo':'Lamborghini'
    }
car_man = pd.Series(car_man_dict)
print(car_man)
```

```python
car = pd.DataFrame({'price':car_price,'manufacturer':car_man})
car
```

```python
car['price']
```
